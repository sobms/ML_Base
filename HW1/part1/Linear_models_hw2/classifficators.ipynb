Логистическая регрессия

```python
import numpy as np
```

```python
class Logistic_Regression:
    def __init__(self, l1_ratio):
        self.l1_ratio = l1_ratio
        self.w = None
    def generate_batches(self, X, y, batch_size):
        np.random.seed(42)
        X = np.array(X)
        y = np.array(y)
        perm = np.random.permutation(len(X))
        for n in range(0,len(X)//batch_size):
            yield (X[perm][n*batch_size:n*batch_size+batch_size],y[perm][n*batch_size:n*batch_size+batch_size])
    
    def fit(self, X, y, num_epochs = 5, lr = 0.03, batch_size = 1000):
        if self.w is None:
            np.random.seed(42)
            self.w = np.random.randn(X.shape[1] + 1)
        
        X_train = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        
        losses = []
        
        for i in range(num_epochs):
            for X_batch, y_batch in self.generate_batches(X_train, y, batch_size):
                predictions = self.sigmoid(X_batch) 
                predictions = np.clip(predictions, 1e-10, 1 - 1e-10)
                losses.append(-np.mean(y_batch * np.log(predictions) + (1 - y_batch) * np.log(1 - predictions)))
                self.w -= self.get_grad(X_batch, y_batch) * lr
        return losses

    def predict(self, X, treshold = 0.5):
        X_ = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        return self.sigmoid(X_)>treshold
    
    def get_grad(self, X, y):
        grad_basic = np.dot(X.T, (self.sigmoid(X) - y))
        grad_l1 =  self.l1_ratio * self.w
        grad_l2 = 1/2*(1-self.l1_ratio) * np.sign(self.w)
        return grad_basic + grad_l1 + grad_l2
    
    def sigmoid(self, X):
        return 1/(1+np.exp((-1)*np.dot(X,self.w)))                    
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
import itertools
```
